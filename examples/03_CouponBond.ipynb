# CouponBond examples

`CouponBond` class is an implementation for fixed coupon bonds. Implemented as a collection of `ZeroCouponBond` objects

```python
from quantpy.instruments.CouponBond import CouponBond
from quantpy import QuantDatetime
```

A `CouponBond` object can be initialized by setting the issuance date, maturity date, coupon rate, notional and coupon frequency

```python
issuance_date = QuantDatetime(2025, 12, 31, calendar = 'London', convention = 'ACT/360')
maturity_date = QuantDatetime(2028, 12, 31, calendar = 'London', convention = 'ACT/360')
bond          = CouponBond(issuance_date, maturity_date, 0.06, notional = 100, coupon_freq = 1)
print(repr(bond))
```

The constituent zeros can be accessed with the `zeros` parameter

```python
bond.zeros
```

Depending on the coupon frequency the number of zeros and their notionals varies

```python
issuance_date = QuantDatetime(2025, 12, 31, calendar = 'London', convention = 'ACT/360')
maturity_date = QuantDatetime(2028, 12, 31, calendar = 'London', convention = 'ACT/360')
bond          = CouponBond(issuance_date, maturity_date, 0.06, notional = 100, coupon_freq = 2)
print(repr(bond))
```

```python
bond.zeros
```

Note that by default the `CouponBond` object makes some assumptions based on the coupon dates. For example, if the maturity date coincides with a month end, the coupon dates are also forced to coincide with a month end
