# QuantDatetime examples

`QuantDatetime` class is a general purpose datetime class with special functionality for handling day-count conventions and calendars. Built on top of the `QuantLib.Date` class.

```python
from quantpy import QuantDatetime
```

A date object can be defined by setting the year, month and day of a `QuantDatetime` object. Note the default time, day-count convention and calendar.

```python
date = QuantDatetime(2025, 12, 31)
print(repr(date))
```

The default values can be defined when initializing a `QuantDatetime` object

```python
date = QuantDatetime(2025, 12, 31, 16, 0, 0, 0, 'London', 'ACT/360')
print(repr(date))
```

A new date object can be created by shifting the date a specified amount using the `shift` method

```python
new_date = date.shift(years=1.)
print(repr(new_date))
```

```python
new_date = date.shift(days=365, hours=2, minutes=3)
print(repr(new_date))
```

The year fraction between two dates can be calculated with the `timedelta` method

```python
date.timedelta(new_date)
```

```python
date.timedelta(date.shift(hours=2, minutes=3))
```

Note that as the used day-count convention is `ACT/360` the year fraction is slightly greater than just one year and correct intraday delta

Dates can be compared with the comparison operators

```python
new_date == date
```

```python
new_date > date
```
