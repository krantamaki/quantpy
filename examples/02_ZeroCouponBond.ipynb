# ZeroCouponBond examples

`ZeroCouponBond` class is an implementation for singular cashflows

```python
from quantpy.instruments.ZeroCouponBond import ZeroCouponBond
from quantpy import QuantDatetime
```

A `ZeroCouponBond` object can be initialized by setting the maturity date and the notional.

```python
maturity_date = QuantDatetime(2026, 12, 31)
zero          = ZeroCouponBond(maturity_date, notional = 100)
print(repr(zero))
```

The zero-coupon bond can be priced with the call method. By default the pricing method is standard discounting

```python
zero(QuantDatetime(2025, 12, 31), 0.03)
```
