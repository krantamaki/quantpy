# Historical index data

```python
import pandas as pd
import refinitiv.data as rd

rd.open_session()
```

## Instructions

1. Find the wanted RICs
2. Fill in the values below

```python
start_date      = '2000-01-01'                                           # Start date of the historical data interval in YYYY-MM-DD format
end_date        = '2025-12-31'                                           # End date of the historical data interval in YYYY-MM-DD format
# Some RICs:
#    .FTEBIG_3AEURT  : FTSE EURO BROAD INVESTMENT-GRADE BOND AAA TOTAL RETURN INDEX EUR END OF DAY
#    .FTEBIG_2AEURT  : FTSE EURO BROAD INVESTMENT-GRADE BOND AA TOTAL RETURN INDEX EUR END OF DAY
#    .FTEBIG_1AEURT  : FTSE EURO BROAD INVESTMENT-GRADE BOND A TOTAL RETURN INDEX EUR END OF DAY
#    .FTEBIG_3BEURT  : FTSE EURO BROAD INVESTMENT-GRADE BOND BBB TOTAL RETURN INDEX EUR END OF DAY
#    .FTEHYM_EUREURT : FTSE EURO HIGH-YIELD BOND TOTAL RETURN INDEX EUR END OF DAY  (BB+ and lower)
#    ITEEU5Y=        : iTraxx Europe
fi_fields        = ["TRDPRC_1", "MOD_DURTN", "MAC_DURTN", "CONVEXITY", "AVG_LIFE", "YLDTOMAT"]
cds_fields       = ["MID_SPREAD"]
# The RICs for the wanted indexes. Will be saved in their own files
fi_index_rics    = [".FTEBIG_3AEURT", ".FTEBIG_2AEURT", ".FTEBIG_1AEURT", ".FTEBIG_3BEURT", ".FTEHYM_EUREURT"]
cds_index_rics   = ["ITEEU5Y="]
save_as          = f'<ric>_{start_date}_to_{end_date}.csv'                # Name of the file where the historical data are stored. The '<ric>' will be replaced with relevant RIC
```

## Run the script

```python
for ric in fi_index_rics:
    ric_df = rd.get_history([ric], fields=fi_fields, start=start_date, end=end_date)
    ric_df.to_csv(save_as.replace("<ric>", ric.replace('.', '')))
```

```python
for ric in cds_index_rics:
    ric_df = rd.get_history([ric], fields=cds_fields, start=start_date, end=end_date)
    ric_df.to_csv(save_as.replace("<ric>", ric.replace('.', '')))
```
