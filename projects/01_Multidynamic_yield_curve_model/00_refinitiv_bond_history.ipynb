# Curve sources

## Instructions

- Use the `<GOVSRCH>` command to open the advanced search for government bonds

- Add columns with coupon info like coupon frequency, next coupon date etc.

- Narrow the search by choosing the issuer, currency and setting the bond grade to investment grade

- Export the query and run it in Codebook environment

```python
import refinitiv.data as rd
import pandas as pd
```

```python
rd.open_session()
```

```python
# Fill the below values
start_date = '2025-12-31'
end_date   = '2026-03-06'
```

```python
def reindex_history(df: pd.DataFrame) -> pd.DataFrame:
    """
    Rearranges the dataframe by turning the top level column names into column values
    """
    top_level_cols = list(df.columns.levels)[0]
    dfs = []
    
    for col in top_level_cols:
        tmp_df        = df[col].copy()
        tmp_df["RIC"] = col
        tmp_df        = tmp_df.reset_index()
        
        dfs.append(tmp_df)
        
    return pd.concat(dfs, ignore_index = True)  
```

## United States

```python
us_bond_df = rd.discovery.search(
	view = rd.discovery.Views.GOV_CORP_INSTRUMENTS,
	query = "Name(United States)",
	top = 1000,
	filter = "((DbType eq 'GOVT' or DbType eq 'CORP' or DbType eq 'AGNC' or DbType eq 'OMUN' or DbType eq 'OTHR') and IsActive eq true and ((FIViewTicker eq 'USTB' or FIViewTicker eq 'GOVT')))",
	select = "RIC,EJVAssetID,DTSubjectName,BusinessEntity,PI,SearchAllCategoryv3,SearchAllCategoryv2,SearchAllCategory,DBSTicker,CouponRate,MaturityDate,IssueDate,ISIN,RCSCurrencyLeaf,RCSCountryLeaf,DbTypeDescription,InstrumentTypeDescription,RCSCouponTypeGenealogy,FaceIssuedUSD,RCSBondGradeLeaf,FIViewTicker"
)
```

```python
us_history_df = rd.get_history(universe = us_bond_df["RIC"].dropna().to_list(), 
                               fields   = ["BID", "ASK", "ACCR_INT", "MID_YLD_1", "CONVEXITY", "MOD_DURTN", "BPV", "ZSPREAD", "OIS_SPREAD"],
                               start    = start_date,
                               end      = end_date
)
us_history_df = reindex_history(us_history_df)
```

```python
us_bond_df.to_csv("us_bonds.csv", index=False)
us_history_df.to_csv("us_history.csv", index=False)
```
